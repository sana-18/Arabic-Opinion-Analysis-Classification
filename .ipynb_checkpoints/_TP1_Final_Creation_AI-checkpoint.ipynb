3) Extraire du texte à partir de HTML avec des méthodes de chaîne pour rechercher les balises
( < div class="comment-text" > )afin de récupérer les commentaires de la page Web: 

#### 1 ere approche:

- recherche l'index de la première occurrence de start_tag qui est'< div class="comment-text" >'

```python
start_tag = '<div class="comment-text">'
start_index = decoded_content.find(start_tag)
```

- recherche l'index de la première occurrence de end_tag, mais commence la recherche à partir de l'index start_index + len(start_tag). Cela permet de trouver la fin de la balise div qui correspond à la balise de début trouvée précédemment. 

```python
end_tag = '</div>'
end_index = decoded_content.find(end_tag, start_index + len(start_tag))
```

- Extraire le texte entre les balises cad les commentaires et les afficher.

```python
# Initialiser l'indice de début
start_index = 0

# Itérer pour extraire tous les commentaires
while start_index != -1:
    # Rechercher la balise de début
    start_index = decoded_content.find(start_tag, start_index)

    # Si la balise de début est trouvée
    if start_index != -1:
        # Trouver la balise de fin à partir de l'index de début
        end_index = decoded_content.find(end_tag, start_index + len(start_tag))

        # Extraire le texte entre les balises
        comment_text = decoded_content[start_index + len(start_tag):end_index]

        # Afficher le texte extrait 
        print(comment_text)

        # Mettre à jour l'indice de début pour la prochaine itération
        start_index = end_index
```

```python
import pandas as pd
# Charger les stopwords à partir d'un fichier
with open('stopwords (1).txt', 'r', encoding='utf-8') as f:
    stopwords_arabic = set(f.read().splitlines())

# Fonction pour supprimer les stopwords en arabe
def supprimer_stopwords(text):
    words = text.split()
    filtered_words = [word for word in words if word.lower() not in stopwords_arabic]
    return ' '.join(filtered_words)

# Appliquer la fonction de suppression de stopwords à la colonne "Commentaire"
df['Comment'] = df['Comment'].apply(supprimer_stopwords)

# Afficher le DataFrame mis à jour
print(df)
```

19) Utiliser la méthode metrics pour calculer les métriques d’évaluation ainsi que la matrice
de confusion.

```python
def metrics(y_train,y_train_pred,y_test,y_test_pred):
    print("Training Accuracy = ", round(accuracy_score(y_train,y_train_pred),2)*100)
    print("\nTrain Classification Report : \n",classification_report(y_train,y_train_pred))
    train_CM = confusion_matrix(y_train,y_train_pred)
    print("\nTrain Confusion Matrix:\n ",train_CM )
    plt.show()
    print("Test Accuracy = ", round(accuracy_score(y_test,y_test_pred),2)*100)
    print("\nTest Classification Report : \n",classification_report(y_test,y_test_pred))
    test_CM = confusion_matrix(y_test,y_test_pred)
    print("\nTest Confusion Matrix:\n ",test_CM)
    plt.show()
```

```python
from nltk import word_tokenize
from nltk.stem.isri import ISRIStemmer

st = ISRIStemmer()
```

```python
def filter_comments(comment):
    #Tokenization
    words = word_tokenize(comment)
    
    #Instancier le stemmer pour la langue arabe
    st = ISRIStemmer()
    
    #Appliquer la racinisation (stemming)
    stemmed_words = [st.stem(word) for word in words]
    
    return stemmed_words

df_f['Filtered_Comments'] = df_f['Commentaire'].apply(filter_comments)
```

```python
df_f['Filtered_Comments'].sample(4)
```

```python
#Sélectionner les colonnes 'Filtered_Comments' et 'Sentiment' du DataFrame df_f
df_f_2 = df_f[['Filtered_Comments', 'Sentiment']]

#Initialiser le vectoriseur TF-IDF
tfidf_2 = TfidfVectorizer()

#Concaténer les mots dans chaque commentaire pour créer une seule chaîne de texte par commentaire
df_f_2['Filtered_Comments'] = df_f_2['Filtered_Comments'].apply(lambda x: ' '.join(x))

#Appliquer la vectorisation TF-IDF aux commentaires
vector_2 = tfidf_2.fit_transform(df_f_2['Filtered_Comments'])

#Extraire les étiquettes (Sentiment) du DataFrame original
y = df_f_2['Sentiment']

#Diviser les données en ensembles d'entraînement et de test
x_train, x_test, y_train, y_test = train_test_split(vector_2, y, test_size=0.2, random_state=42)

#Initialiser le classificateur SVM avec un noyau linéaire
svm = SVC(kernel='linear')

#Entraîner le modèle SVM sur l'ensemble d'entraînement
svm.fit(x_train, y_train)

#Prédire les étiquettes sur l'ensemble de test et d'entraînement
y_test_pred = svm.predict(x_test)
y_train_pred = svm.predict(x_train)

#Évaluer les performances du modèle en utilisant la fonction metrics 
metrics(y_train, y_train_pred, y_test, y_test_pred)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

# Créer un CountVectorizer
count_vectorizer = CountVectorizer()
count_matrix = count_vectorizer.fit_transform(df_f['Commentaire'])

# Calculer la somme de chaque mot sur l'ensemble des commentaires
word_frequencies = count_matrix.sum(axis=0)

# Calculer le nombre total de documents
total_docs = count_matrix.shape[0]

# Définir un seuil de fréquence (par exemple, 0.2 pour 20%)
threshold = 0.1

# Trouver les mots à supprimer (ceux qui apparaissent dans plus de 20% des documents)
words_to_remove = set()
for word, frequency in zip(count_vectorizer.get_feature_names_out(), (word_frequencies / total_docs).tolist()[0]):
    if frequency >=threshold:
        words_to_remove.add(word)

# Créer un nouveau vocabulaire en supprimant les mots moins pertinents
filtered_vocab = [word for word in count_vectorizer.get_feature_names_out() if word not in words_to_remove]

# Filtrer les commentaires en supprimant les mots moins pertinents
df_f['Commentaire'] = df_f['Commentaire'].apply(lambda x: ' '.join([word for word in x.split() if word not in words_to_remove]))

# Afficher les mots moins pertinents supprimés
print("Mots moins pertinents supprimés:")
print(words_to_remove)

# Afficher le nouveau vocabulaire
print("\nNouveau vocabulaire:")
print(filtered_vocab)
print(len(filtered_vocab))

# Afficher le score de chaque mot
print("\nScore de chaque mot:")
for word, frequency in zip(count_vectorizer.get_feature_names_out(), (word_frequencies / total_docs).tolist()[0]):
    print(f"{word}: {frequency}")
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Assuming x_train, y_train, x_test, y_test are already defined

# Create a KNN classifier
knn = KNeighborsClassifier(n_neighbors=5)  # You can adjust the number of neighbors (n_neighbors) as needed

# Train the KNN model on the training set
knn.fit(X_train, y_train)

# Predict labels on the test and training sets
y_test_pred_knn = knn.predict(W_test)
y_train_pred_knn = knn.predict(X_train)

# Evaluate the performance of the KNN model using metrics
def metrics(y_train, y_train_pred, y_test, y_test_pred):
    print("Training Accuracy:", accuracy_score(y_train, y_train_pred))
    print("Testing Accuracy:", accuracy_score(y_test, y_test_pred))
    print("\nConfusion Matrix (Test Set):\n", confusion_matrix(y_test, y_test_pred))
    print("\nClassification Report (Test Set):\n", classification_report(y_test, y_test_pred))

# Evaluate the performance of the KNN model
metrics(y_train, y_train_pred_knn, y_test, y_test_pred_knn)
```
